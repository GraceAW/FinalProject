```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
gap_url = 'https://raw.githubusercontent.com/chendaniely/pandas_for_everyone/master/data/gapminder.tsv'
```

```python
df = pd.read_csv(gap_url, sep ='\t')
```

```python
df
```

```python
fig, ax= plt.subplots(figsize=(12, 6))
df.plot.scatter(x = 'gdpPercap', y = 'lifeExp', c='year', colormap='viridis', logx=True, ax=ax)

plt.title('Scatter Plot to Visualize LifeExp with Respect to GdpPercap and Year')

plt.show()
```
